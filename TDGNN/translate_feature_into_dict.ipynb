```python
import pickle
import numpy as np
```

```python
datapath = './ia-conact/feature_random_contact.txt'
res = []
with open(datapath) as f:
    for line in f:
        temp = []
        tokens = line.strip().split()
        for nums in tokens:
            temp.append(float(nums))
        res.append(temp)
```

```python
res[0][0]
```

```python
feature_vec = dict()
```

```python
for item in res:
    feature_vec[int(item[0])] = np.array(item[1:])
```

```python
len(feature_vec[256])
```

```python
with open('./ia-conact/feature_vec', 'wb') as f:
    pickle.dump(feature_vec, f)
```
